```python

# Createing the preprocessed images folder if it doesn't exist
os.makedirs(preprocessed_folder, exist_ok=True)

# Defining target size for EfficientNet-B4
target_size = (380, 380)

# Function to preprocess images
def preprocess_image(image_path, target_size):
    img = load_img(image_path, target_size=target_size)
    img_array = img_to_array(img)
    preprocessed_img = preprocess_input(img_array)
    return img_array, preprocessed_img  

# Preprocessing and saving all images
for image_id in df['image_id']:
    image_path = os.path.join(image_folder, image_id + ".jpg")
    
    if not os.path.exists(image_path):  
        print(f"Warning: {image_path} not found. Skipping.")
        continue  
    
    original_img, preprocessed_img = preprocess_image(image_path, target_size)
    
    save_path = os.path.join(preprocessed_folder, image_id + ".jpg")
    save_img(save_path, original_img) 
    

print("All images have been preprocessed and saved.")
```

```python


# 1. Set seeds for full reproducibility
seed = 42
random.seed(seed)
np.random.seed(seed)
tf.random.set_seed(seed)

# 2. Paths and data
df = pd.read_csv("C:/Users/styvo/Desktop/skin lesion/HAM10000_metadata.csv")
image_folder = "C:/Users/styvo/Desktop/skin lesion/HAM10000_preprocessed_images_efficient net"

# Label Encoding
label_encoder = LabelEncoder()
df['label_encoded'] = label_encoder.fit_transform(df['dx'])

# One-hot encode
num_classes = df['label_encoded'].nunique()
one_hot_labels = to_categorical(df['label_encoded'], num_classes=num_classes)
df['one_hot_labels'] = list(one_hot_labels)

target_size = (380, 380)

# 3. Load and preprocess image
def load_and_preprocess_image(image_id, label):
    image_path = tf.strings.join([image_folder, image_id + ".jpg"], separator='/')
    image = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, target_size)
    image = preprocess_input(image)
    return image, label

# 4. Simple augmentation inside tf.data
def augment(image, label):
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_flip_up_down(image)
    image = tf.image.random_brightness(image, max_delta=0.15)
    image = tf.image.random_contrast(image, lower=0.85, upper=1.15)
    return image, label

# 5. Create dataset
def create_dataset(dataframe, batch_size=32, shuffle=True, augment_data=False):
    image_ids = dataframe['image_id'].values
    labels = np.stack(dataframe['one_hot_labels'].values)
    dataset = tf.data.Dataset.from_tensor_slices((image_ids, labels))
    dataset = dataset.map(load_and_preprocess_image, num_parallel_calls=tf.data.AUTOTUNE)
    if shuffle:
        dataset = dataset.shuffle(buffer_size=len(dataframe))
    if augment_data:
        dataset = dataset.map(augment, num_parallel_calls=tf.data.AUTOTUNE)
    return dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)

batch_size = 16

# Stratified split
train_df, temp_df = train_test_split(df, test_size=0.2, stratify=df['label_encoded'], random_state=seed)
val_df, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df['label_encoded'], random_state=seed)

train_dataset = create_dataset(train_df, batch_size=batch_size, shuffle=True, augment_data=True)
val_dataset = create_dataset(val_df, batch_size=batch_size, shuffle=False, augment_data=False)
test_dataset = create_dataset(test_df, batch_size=batch_size, shuffle=False, augment_data=False)

# 6. Custom Callback to Save History After Every Epoch
class SaveHistory(Callback):
    def __init__(self, file_path="efficientnetb4_history_running.pkl"):
        super().__init__()
        self.file_path = file_path
        self.history = {}

    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        for k, v in logs.items():
            self.history.setdefault(k, []).append(v)
        with open(self.file_path, "wb") as f:
            pickle.dump(self.history, f)

# 7. Build Model
base_model = EfficientNetB4(weights="imagenet", include_top=False, input_shape=(380, 380, 3))
base_model.trainable = False  # Freeze base for now

x = GlobalAveragePooling2D()(base_model.output)
x = Dense(256, activation="relu")(x)
x = Dropout(0.4)(x)
x = Dense(128, activation="relu")(x)
x = Dropout(0.3)(x)
output_layer = Dense(num_classes, activation="softmax")(x)

model = Model(inputs=base_model.input, outputs=output_layer)

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-4),  # LOWERED learning rate
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

# 8. Callbacks
checkpoint = ModelCheckpoint(
    "best_model_accuracy.keras",
    monitor="val_accuracy",
    save_best_only=True,
    mode="max",
    verbose=1
)
early_stop = EarlyStopping(
    monitor="val_loss",
    patience=7,
    restore_best_weights=True
)
reduce_lr = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.2,
    patience=4,
    min_lr=1e-6
)
save_history_callback = SaveHistory("efficientnetb4_history_running.pkl")

# 9. Train model
initial_epochs = 50
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=initial_epochs,
    callbacks=[checkpoint, early_stop, reduce_lr, save_history_callback]
)

# 10. Save Final Training History
with open("efficientnetb4_final_training_history.pkl", "wb") as f:
    pickle.dump(history.history, f)
```

### Grad-Cam Heatmap

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
import cv2

# Load your model
model = load_model("best_model_accuracy.keras")
input_tensor = tf.keras.Input(shape=(380, 380, 3), name="input_layer_2")


# Pick the last conv layer in EfficientNetB4
last_conv_layer_name = "top_conv"

# Get an image and label from the test dataset
for images, labels in test_dataset.take(1):  # Just take 1 batch
    image = images[0]  # first image in batch
    true_label = tf.argmax(labels[0]).numpy()
    break

# Expand dims to add batch size = 1
img_array = tf.expand_dims(image, axis=0)

# Get the model components
grad_model = tf.keras.models.Model(
    [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
)

# Compute gradients
with tf.GradientTape() as tape:
    conv_outputs, predictions = grad_model(img_array)
    pred_index = tf.argmax(predictions[0])
    class_channel = predictions[:, pred_index]

# Get gradient of top predicted class with respect to conv layer output
grads = tape.gradient(class_channel, conv_outputs)

# Compute guided gradients
pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
conv_outputs = conv_outputs[0]
heatmap = tf.reduce_sum(tf.multiply(pooled_grads, conv_outputs), axis=-1)

# Normalize heatmap to [0, 1]
heatmap = np.maximum(heatmap, 0)
heatmap /= tf.math.reduce_max(heatmap)
heatmap = heatmap.numpy()

# Resize heatmap to image size
heatmap = cv2.resize(heatmap, (image.shape[1], image.shape[0]))
heatmap = np.uint8(255 * heatmap)

# Apply heatmap to original image
heatmap_color = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
superimposed_img = heatmap_color * 0.4 + image.numpy()

# Plot
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.imshow(image.numpy().astype("uint8"))
plt.title(f"Original\nTrue: {label_encoder.classes_[true_label]}")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(heatmap, cmap="jet")
plt.title("Grad-CAM Heatmap")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(superimposed_img.astype("uint8"))
plt.title("Overlay")
plt.axis("off")

plt.tight_layout()
plt.show()
```
